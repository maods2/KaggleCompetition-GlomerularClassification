```python
import sys
sys.path.append('..')
import os
# PROJECT_ROOT = os.path.dirname()

from pathlib import Path
import json
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf 

```

```python
def load_results():
    path = Path('../artifacts/')

    experiments = []
    for i in path.glob('*'):
        try:
            new_path = str(i) + '/metrics.json'
            f = open(new_path)
            data = json.load(f)
            experiments.append(data)
        except  Exception as e:
            print(e)
        finally:
            f.close()

    df = pd.DataFrame(experiments)
    df = df.sort_values(by='acc_test', ascending=False)
    return df


def load_crossvalidation_results():
    path = Path('../artifacts/')

    experiments = []
    for i in path.glob('*_CV_*'):
        for j in i.glob('*'):
            try:
  
                new_path = str(j) + str(Path('/metrics.json'))
                f = open(new_path)
                data = json.load(f)
                experiments.append(data)
            except  Exception as e:
                print(e)
            finally:
                f.close()

    if experiments:
        df = pd.DataFrame(experiments)
        df = df.sort_values(by='acc_test', ascending=False)
        return df
    else:
        return pd.DataFrame({})


```
